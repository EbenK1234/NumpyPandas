```python
# 3h. Concatenate two data frames Ghana cars and Japanes Cars
```

```python
GhanaCars = {'Company': ['KANTAKA', 'VW', 'BMW', 'NEOPLAN'], 'Price': [23845,171995,135925,71400]}

JapaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi'], 'Price':[29995,23600,61500,58900]}
GCar = pd.DataFrame.from_dict(GhanaCars)
JCar = pd.DataFrame(JapaneseCars)

df_cars = pd.concat([GCar,JCar], keys=['Ghana','Japan'])
df_cars
```
